#### Load the `employees` dataset into a dataframe. Call the dataframe `employees`.

```python
# your code here
emp_file = "employees.csv"
dpt_file = "departements.csv"

employees = pd.read_csv(emp_file)
departments = pd.read_csv(dpt_file)
employees
```

#### Explore the data types of the `employees` dataframe. Comment your results.

```python
# your code here
employees.info()
```

```python
departments.info()
```

```python
"""
your comments here
"""

employees_dpt = employees.merge(departments, how="left", on="department_id")
employees_dpt
```

#### What's the average salary in this company?

```python
# your code here
employees_dpt.salary.mean()
```

#### What's the highest salary?

```python
# your code here
employees_dpt.salary.max()
```

#### What's the lowest salary?

```python
# your code here
employees_dpt.salary.min()
```

#### Who are the employees with the lowest salary?

```python
# your code here
employees_dpt[employees_dpt.salary == employees_dpt.salary.min()]
```

#### Find all the information about an employee called David.

```python
# your code here
employees_dpt[employees_dpt.first_name == "David"]
```

#### Could you return only David's salary?

```python
# your code here
employees_dpt.salary[employees_dpt.first_name == "David"]
```

#### Find the employees whose title is associate and whose salary is above 55.

#### Group the employees by number of years of employment. What are the average salaries in each group?

```python
# your code here
employees_dpt["hire_year"] = [date[:4] for date in employees_dpt.hire_date]
employees_dpt[["hire_year", "salary"]].groupby("hire_year").mean()
```

####  What is the average salary per title?

```python
# your code here
employees_dpt[["job_id", "salary"]].groupby("job_id").mean()
```

####  Find the salary quartiles.


```python
# your code here
employees_dpt.salary.quantile([0.25, 0.5, 0.75])
```

#### Is the mean salary different per gender?

```python
# your code here
employees_dpt[["department_name", "salary"]].groupby("department_name").mean()
```

#### Find the minimum, mean and maximum of all numeric columns for each company department.



```python
# your code here

employees_dpt_num = employees_dpt.select_dtypes(include = "number")

employees_dpt.groupby("department_name").describe()
```

#### Bonus Question:  for each department, compute the difference between the maximum and the minimum salary.
**Hint**: try using `agg` or `apply` combined with `lambda` functions.

```python
# your code here
pd.DataFrame(employees_dpt.groupby("department_name").max().salary - employees_dpt.groupby("department_name").min().salary)
```
